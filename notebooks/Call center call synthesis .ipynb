```python
import os
import librosa
import numpy as np
import random
import soundfile as sf
```

```python
# synthetic files location
operators_dir = "../data/synthetic_call_center/operators"
customers_dir = "../data/synthetic_call_center/customers"
target_dir = "../data/synthetic_call_center/synthetic_calls2"
```

```python
# for each customer create two calls - each one with different operator
for customer in os.listdir(customers_dir):
    customer_path = os.path.join(customers_dir, customer)
    # choose two operators
    operators = random.sample(os.listdir(operators_dir), 2)
    # for each call choose random utternaces of the samples
    all_customer_utternaces = os.listdir(customer_path)
    customers_utterances_list=[]
    customers_utterances_list.append(random.sample(all_customer_utternaces, round(len(all_customer_utternaces)/2)))
    customers_utterances_list.append(list(set(all_customer_utternaces)-set(customers_utterances_list[0])))
    call_id = 1

    for operator, customer_utterances_list in zip(operators, customers_utterances_list):
        operator_path = os.path.join(operators_dir, operator)
        print("creating a call using:",customer, operator)
        # choose utternaces for the operator side
        oper_utterances_list = random.sample(os.listdir(operator_path), round(len(os.listdir(operator_path))/2))
        
        # build conversation
        is_first_utt = 1
        for utterance_idx in range(10):
            # start from the operator
            if is_first_utt:
                # first utterance should be longer than 5 sec
                op_utter, sr = librosa.core.load(os.path.join(operator_path, "opener.wav"), sr=8000)
                new_call = op_utter
                is_first_utt = 0
            else:
                op_utter, sr = librosa.core.load(os.path.join(operator_path, oper_utterances_list[utterance_idx]), sr=8000)
                new_call = np.append(new_call, op_utter)
            cu_utter, sr = librosa.core.load(os.path.join(customer_path, customer_utterances_list[utterance_idx]), sr=8000)
            new_call = np.append(new_call, cu_utter)

        # Write out audio as 16bit PCM WAV
        sf.write(os.path.join(target_dir,customer+"_"+str(call_id))+".wav", new_call, sr, subtype='PCM_16')
        del new_call
        call_id += 1
        print("done!")
            
```
